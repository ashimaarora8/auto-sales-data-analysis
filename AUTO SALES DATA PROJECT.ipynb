```python
monthly_sales = df.groupby("MONTH")["SALES"].sum().reset_index()
yearly_sales = df.groupby("YEAR")["SALES"].sum().reset_index()
product_sales = df.groupby("PRODUCTLINE")["SALES"].sum().reset_index()
dealsize_sales = df.groupby("DEALSIZE")["SALES"].sum().reset_index()
status_sales = df.groupby("STATUS")["SALES"].sum().reset_index()
country_sales = df.groupby("COUNTRY")["SALES"].sum().reset_index()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(16,6))
plt.suptitle("TIME-BASED SALES ANALYSIS", fontsize=16, fontweight="bold")

# 1. Monthly Sales
plt.subplot(1,2,1)
sns.barplot(x="MONTH", y="SALES", data=monthly_sales)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")

# 2. Yearly Sales
plt.subplot(1,2,2)
sns.barplot(x="YEAR", y="SALES", data=yearly_sales)
plt.title("Yearly Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.suptitle("PRODUCT & CUSTOMER BEHAVIOUR", fontsize=16, fontweight="bold")

# 1. Product Line Sales
plt.subplot(1,2,1)
sns.barplot(x="PRODUCTLINE", y="SALES", data=product_sales)
plt.title("Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

# 2. Deal Size Sales
plt.subplot(1,2,2)
sns.barplot(x="DEALSIZE", y="SALES", data=dealsize_sales)
plt.title("Sales by Deal Size")
plt.xlabel("Deal Size")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.suptitle("OPERATIONAL & GEOGRAPHICAL ANALYSIS", fontsize=16, fontweight="bold")

# 1. Order Status Analysis
plt.subplot(1,2,1)
sns.barplot(x="STATUS", y="SALES", data=status_sales)
plt.title("Sales by Order Status")
plt.xlabel("Status")
plt.ylabel("Total Sales")
plt.xticks(rotation=40)

# 2. Top 10 Countries by Sales
plt.subplot(1,2,2)
sns.barplot(x="COUNTRY", y="SALES", data=country_sales.nlargest(10, "SALES"))
plt.title("Top 10 Countries by Sales")
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
